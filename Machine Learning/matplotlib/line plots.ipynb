```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use("default")

linear_data = np.array([1, 2, 3, 4, 5])
quadratic_data = linear_data**2

plt.plot(linear_data, "-o", quadratic_data, "-o");
```

```python
plt.plot(linear_data, quadratic_data);
```

```python
plt.plot(linear_data, "-o", quadratic_data, "-o")
plt.plot([5, 9, 21], "--r");
```

```python
plt.plot(linear_data, "-o", quadratic_data, "-o")
plt.plot([5, 9, 21], "--r")

plt.xlabel("label X")
plt.ylabel("label Y")
plt.title("Title")

plt.legend(["linear", "quadratic", "random"]);
```

```python
plt.plot(linear_data, "-o", quadratic_data, "-go")
plt.plot([5, 9, 21], "--r")

plt.gca().fill_between(range(len(linear_data)), 
                       linear_data, quadratic_data,
                      facecolor="yellow",
                      alpha=0.5);
```
