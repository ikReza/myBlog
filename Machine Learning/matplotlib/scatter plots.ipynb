## Scatterplots

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = x

plt.scatter(x, y);
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = x
colors = []
for i in range(len(x)):
    colors.append("green") if (i%2 == 0) else colors.append("red")

plt.scatter(x, y, s=100, c=colors)
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = x
colors = ["green"]*(len(x)-1)
colors.append("red")

plt.scatter(x, y, s=100, c=colors)
plt.show()
```

### zip unpacking

For unpacking <span style="color: tomato">zip</span>, we can use <span style="color: tomato">*</span>

```python
zip_generator = zip([1, 2, 3, 4, 5],[5, 4, 3, 2, 1])
x, y = zip(*zip_generator)
print(x)
print(y)
```

```python
plt.scatter(x[:2], y[:2], c="tomato", s=100, label="tall student")
plt.scatter(x[2:], y[2:], c="blue", s=100, label="short student")
```

```python
plt.scatter(x[:2], y[:2], c="tomato", s=100, label="tall student")
plt.scatter(x[2:], y[2:], c="blue", s=100, label="short student")

plt.xlabel("something on x label")
plt.ylabel("something on y label")
plt.title("Checking the title")

plt.legend()
```

```python
plt.scatter(x[:2], y[:2], c="tomato", s=100, label="tall student")
plt.scatter(x[2:], y[2:], c="blue", s=100, label="short student")

plt.xlabel("something on x label")
plt.ylabel("something on y label")
plt.title("Checking the title")

plt.legend(loc=3, title="Legend", frameon=False);
```

##### loc and frameon

<span style="color: red">loc</span> defines the location where we want to display the legend

- loc=1 *top-right*
- loc=2 *top-left*
- loc=3 *bottom-left*
- loc=4 *bottom-right*

anti-clockwise rotation

<span style="color: red">frameon</span> defines do we want boundary around legend or not
