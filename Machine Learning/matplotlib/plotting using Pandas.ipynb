```python
df = pd.read_csv("car-sales.csv")
df.head()
```

##### We want to remove the ($) sign and commas from the Price column

```python
df["Price"] = df["Price"].str.replace("[\$\,\.]", "")
df.head()
```

```python
type(df["Price"][0])
```

```python
car_sales = df.copy()
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales.head()
```

##### We want to add an extra column - sale date

```python
car_sales["Sale Date"] = pd.date_range("8/19/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales.head()
```

##### Drawing with pandas dataframe

```python
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
