```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import style
style.use("default")
```

```python
food_prices = {"coke": 100, "cake": 350, "cookie": 240}

fig, ax = plt.subplots()
ax.bar(food_prices.keys(), food_prices.values())
ax.set(title="Food store ❤",
      ylabel="Price (£)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(food_prices.keys()), list(food_prices.values()))
ax.set(title="Food store ❤",
      xlabel="Price (£)");
```
