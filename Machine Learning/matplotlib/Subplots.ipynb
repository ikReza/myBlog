```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import style
style.use("default")
```

```python
x = np.random.randn(1000)
food_prices = {"coke": 100, "cake": 350, "cookie": 240}
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5));
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5));

ax1.plot(x, x/2)
ax1.set(title="Line plot")

ax2.scatter(np.random.rand(10), np.random.rand(10))
ax2.set(title="scatter plot")

ax3.bar(food_prices.keys(), food_prices.values())
ax3.set(title="Bar chart")

ax4.hist(np.random.randn(1000))
ax4.set(title="Histogram");
```
