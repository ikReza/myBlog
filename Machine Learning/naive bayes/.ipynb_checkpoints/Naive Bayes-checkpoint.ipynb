```python
from sklearn.datasets import fetch_20newsgroups
```

```python
data = fetch_20newsgroups()
```

```python
data.keys()
```

```python
print(data.DESCR)
```

```python
categories = data.target_names
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
train = fetch_20newsgroups(subset="train", categories=categories)
test = fetch_20newsgroups(subset="test", categories=categories)
```

```python
print(len(train.data), len(test.data))
round((len(test.data)/(len(test.data)+len(train.data)))*100)
```

```python
X_train, X_test, y_train, y_test = (train.data, test.data, train.target, test.target)
```

#### import necessary packages

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

#### create model

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

#fit the model
model.fit(X_train, y_train)

#prediction
y_pred = model.predict(X_test)
```

#### confusion matrix

```python
from sklearn.metrics import confusion_matrix
from matplotlib import style
style.use("default")
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm.T, square=True, annot=True, fmt="d", cbar=False, 
            xticklabels=train.target_names, yticklabels=train.target_names)

#plotting
plt.xlabel("True label")
plt.ylabel("Predicted label")
plt.autoscale()
plt.show()
```

#### Predicting category based on our training

```python
def predict_category(s, train=train, model=model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
predict_category("Enjoy your life to the fullest.")
```

## Another example 

```python
import pandas as pd
df = pd.read_csv("spam.csv")
df.head(10)
```

```python
trial = df.copy()
print(trial.columns)
trial.drop(["Unnamed: 0", "label_num"],axis = 1, inplace = True)
trial.groupby("label").describe()
```

```python
X = df.text
y = df.label_num
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
emails = ["Hello, nice to meet you. How you doing these days",
         "Upto 20% discount on parking, exclusive offer just for you. Don't miss this reward",
         "We're going to Saint Martin today. Let's go"]
```
