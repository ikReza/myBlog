```python
import networkx as nx
import matplotlib.pyplot as plt
#to remove matplotlib warning
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
#G = nx.Graph()
G = nx.DiGraph()

edges = [("a", "b", 10), ("a", "c", 3), 
         ("b", "c", 1), ("b", "d", 2), 
         ("c", "b", 4), ("c", "d", 8), ("c", "e", 2),
         ("d", "e", 7),
         ("e", "d", 9)]

G.add_weighted_edges_from(edges)
```

```python
#shortest path
shortestPath = nx.dijkstra_path(G, "a", "d")
print(f"Shortest path -> {shortestPath}")

#shortest distance
shortest_distance = nx.dijkstra_path_length(G, "a", "d")
print(shortest_distance)

# number of nodes and edges
print(G.number_of_edges(), G.number_of_nodes())
```

```python
#visualize graph
#Position of nodes using Fruchterman-Reingold force-directed algorithm.
pos = nx.spring_layout(G)
weight = nx.get_edge_attributes(G, "weight")
nx.draw_networkx(G, pos, arrow=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=weight)
plt.show()
```
